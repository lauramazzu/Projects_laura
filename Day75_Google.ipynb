```python
#create locators for ticks on the time axis
years= mdates.YearLocator()
months=mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')


#increases size and resolution
#this part goes always first
plt.figure(figsize=(14,8), dpi = 120)
plt.title('Tesla Web Search vs Price', fontsize = 18)

#increase fontsize and linewidth for larger charts, and rotate xaxis
#this part goes always first
plt.xticks(fontsize=14)
plt.yticks(fontsize = 14)


ax1 = plt.gca() #get the axis
ax2 = ax1.twinx() #create another axis that shares the same x-axis

#change font of secondary axis
ax2.tick_params(labelsize = 14)



# Rotating X-axis labels
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)


#
#ticksax1.xaxis.set_major_locator(years)ax1.xaxis.set_major_formatter(years_fmt)ax1.xaxis.set_minor_locator(months)


ax1.set_xlabel('Year', fontsize=18)
ax1.set_ylabel('Stock Price', fontsize = 18, color = '#FF5733')
ax2.set_ylabel('Search trend', fontsize = 18, color = 'skyblue')

#Set the minimum and maximum values on the axes
ax1.set_ylim([0, 600])
ax1.set_xlim([df_tesla.MONTH.min(), df_tesla.MONTH.max()])


ax1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE, '#FF5733', linewidth = 3)
ax2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, 'skyblue', linewidth =3)



#displayschart explicitly
plt.show()
```

```python
#create locators for ticks on the time axis
years= mdates.YearLocator()
months=mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')


#increases size and resolution
#this part goes always first
plt.figure(figsize=(14,8), dpi = 120)
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate', fontsize = 18)
plt.grid(linestyle = '--')


#increase fontsize and linewidth for larger charts, and rotate xaxis
#this part goes always first
plt.xticks(fontsize=14)
plt.yticks(fontsize = 14)


ax1 = plt.gca() #get the axis
ax2 = ax1.twinx() #create another axis that shares the same x-axis

#change font of secondary axis
ax2.tick_params(labelsize = 14)



# Rotating X-axis labels
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)


# format the ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)


ax1.set_xlabel('Year', fontsize=18)
ax1.set_ylabel('FRED U/E RATE', fontsize = 18, color = '#0B5505')
ax2.set_ylabel('Search trend', fontsize = 18, color = '#4F0555')

#Set the minimum and maximum values on the axes
ax1.set_ylim([3, 10.5])
#ax1.set_xlim([df_unemployment_one.MONTH.min(), df_unemployment_one.MONTH.max()])

#calculate the rolling average over a 6month window
roll_df = df_unemployment_one[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=3).mean()




ax1.plot(df_unemployment_one.MONTH[:-24], roll_df.UNRATE[:-24], '#0B5505', linewidth = 3,  linestyle = 'dashed')
ax2.plot(df_unemployment_one.MONTH[:-24], roll_df.UE_BENEFITS_WEB_SEARCH[:-24], '#4F0555', linewidth =3)



#displayschart explicitly
plt.show()
```
