```python
prize_year = df_clean.groupby('year').count()
prize_year_ = prize_year.prize
prize_year_
```

```python
moving_average = prize_year_.rolling(window = 5).mean()
```

```python
plt.figure(figsize=(16,10))
plt.title('Number of Prizes Awarded per Year', fontsize = 20)

#xticks = np.linspace(1900, 2020, 24)

plt.xticks(ticks = np.arange(1900, 2021, step = 5), fontsize=14, rotation = 45)
plt.yticks(fontsize=14)

# Rotating X-axis labels
#for tick in xticks:
#    tick.set_rotation(45)


plt.xlabel('Year', fontsize=18)
plt.ylabel('Prizes', fontsize = 18)

plt.ylim(0, 15)

plt.scatter(prize_year_.index, prize_year_.values, color ="darkcyan", alpha = 0.7, s = 100)

plt.plot(prize_year_.index, moving_average.values, c = "crimson", linewidth = 3)

plt.show()
```

```python
#are more prices shared than before?
#average price share of the winners on a year by yearbasis
yearly_avg_share = df_clean.groupby('year').agg({'share_pct': pd.Series.mean})
share_moving_average = yearly_avg_share.rolling(window = 5).mean()

print(yearly_avg_share)
share_moving_average
```

```python
#visualise the data
plt.figure(figsize=(16,8), dpi=200)
plt.title('Number of Nobel Prizes Awarded per Year', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(ticks=np.arange(1900, 2021, step=5),
           fontsize=14,
           rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx() # create second y-axis
ax1.set_xlim(1900, 2020)

ax1.scatter(x=prize_year_.index,
           y=prize_year_.values,
           c='dodgerblue',
           alpha=0.7,
           s=100,)

ax1.plot(prize_year_.index,
        moving_average.values,
        c='crimson',
        linewidth=3,)

# Adding prize share plot on second axis
ax2.plot(prize_year_.index,
        share_moving_average.values,
        c='grey',
        linewidth=3,)

plt.show()
```

```python
#Invert the secondary y axis to see the relationship between the number of prizes and laureate share

#visualise the data
plt.figure(figsize=(16,8), dpi=200)
plt.title('Number of Nobel Prizes Awarded per Year', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(ticks=np.arange(1900, 2021, step=5),
           fontsize=14,
           rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx() # create second y-axis
ax1.set_xlim(1900, 2020)

#invert axis
ax2.invert_yaxis()

ax1.scatter(x=prize_year_.index,
           y=prize_year_.values,
           c='dodgerblue',
           alpha=0.7,
           s=100,)

ax1.plot(prize_year_.index,
        moving_average.values,
        c='crimson',
        linewidth=3,)

# Adding prize share plot on second axis
ax2.plot(prize_year_.index,
        share_moving_average.values,
        c='grey',
        linewidth=3,)

plt.show()
```

```python
#we can count the number of prizes by country by year
prize_by_year = df_clean.groupby(['birth_country_current', 'year'], as_index = False).count()
prize_by_year
prize_by_year = prize_by_year.sort_values('year')[['year', 'birth_country_current', 'prize']]
prize_by_year
```

```python
#we can create a series that has the cumulative sum for the number of prizes won
cumulative_prizes = prize_by_year.groupby(['birth_country_current', 'year']).sum().groupby(level = [0]).cumsum()
cumulative_prizes.reset_index(inplace = True)
cumulative_prizes
```

```python
l_chart = px.line(cumulative_prizes,
                  x = 'year',
                  y = 'prize',
                  color = 'birth_country_current',
                  hover_name = 'birth_country_current',
                  )

l_chart.update_layout(xaxis_title='Year',
                      yaxis_title='Number of Prizes')

l_chart.show()
```

```python
#calculate the age of the laureate in the year of the ceremony
from posixpath import expanduser

#birth_date = df_clean.birth_date.to_string()

birth_year = df_clean.birth_date.dt.year


df_clean['winning_age'] = df_clean.year - birth_year
df_clean
```

```python
#Using Seaborn to creating an histogram to visualise the distribution of laureate age
#at the time of winning

histo_plot = sns.histplot(data = df_clean, x = df_clean.winning_age.values, kde = True, color = '#99d8c9', bins = 10)
histo_plot.set(xlabel = 'Years of Nobel Prize Winners')

plt.title('Distribution of Age on Receipt of Prize')
plt.show
```

```python
#Are the Nobel laureates being nominated later in life than before?
#Age of the laureate at the time of the award increased oor diecreased over time

plt.figure(figsize=(8,4), dpi = 200)
with sns.axes_style("whitegrid"):
  sns.regplot(data= df_clean, x="year", y = "winning_age", lowess = True, scatter_kws = {'alpha': 0.4}, line_kws = {'color': 'black'})

plt.show()
```

```python
#How does the age of laureates vary by category?
plt.figure(figsize=(8,4), dpi = 200)
with sns.axes_style("whitegrid"):
  sns.boxplot(data= df_clean, x= 'category', y= 'winning_age', palette = "cubehelix")

plt.show()
```

```python
#Use Seaborn .lmplot() to create 6 separate charts for each prize category
plt.figure(figsize=(8,4), dpi = 200)
with sns.axes_style("whitegrid"):
  sns.lmplot(data = df_clean, x= "year", y= "winning_age", row = "category",lowess= True, aspect = 2, scatter_kws = {'alpha':0.6}, line_kws = {'color': 'black'}, palette = 'Blues' )

plt.show()
```

```python
#Combine all these charts into the same chart with the hue parameter
with sns.axes_style("whitegrid"):
    sns.lmplot(data=df_clean,
               x='year',
               y='winning_age',
               hue='category',
               lowess=True,
               aspect=2,
               scatter_kws={'alpha': 0.5},
               line_kws={'linewidth': 5},
               palette = "Accent")

plt.show()
```
